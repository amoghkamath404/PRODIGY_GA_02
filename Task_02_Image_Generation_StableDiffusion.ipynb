```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

pipe = pipe.to("cuda")

prompt = "A hyper-realistic close-up of a futuristic CPU chip with glowing fiber-optic circuitry, copper and gold textures, microscopic detail, Rembrandt lighting, dark moody background, 8k resolution, cinematic atmosphere, macro photography style"

image = pipe(prompt).images[0]

image.save("generated_image.png")

image.show()
```
